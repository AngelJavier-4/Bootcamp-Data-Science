```python
import pandas as pd
import numpy as np

#Lectura de los csv
RecordsWinningStreaks = pd.read_csv('RecordsWinningStreaks.csv')
RecordsMostMatchesPlayed = pd.read_csv('RecordsMostMatchesPlayed.csv')
RecordsMostMatchesWon = pd.read_csv('RecordsMostMatchesWon.csv')
RecordsGreatestWinningPct = pd.read_csv('RecordsGreatestWinningPct.csv')
RecordsATPRanking = pd.read_csv('RecordsATPRanking.csv')
RecordsAllRecords = pd.read_csv('RecordsAllRecords.csv')
```

```python
#Retocada la tabla de Records Winning Streaks
RWS = RecordsWinningStreaks.copy()
RWS.rename(columns = {'value':'Winning Streak', 'recordHolders_0_name':'Record Holder 1', 'recordHolders_0_country_id':'Country 1',
                     'recordHolders_0_detail':'Detail 1', 'recordHolders_1_name':'Record Holder 2', 'recordHolders_1_country_id':'Country 2',
                     'recordHolders_1_detail':'Detail 2', 'recordHolders_2_name':'Record Holder 3', 'recordHolders_2_country_id':'Country 3',
                     'recordHolders_2_detail':'Detail 3'}, inplace = True)

del RWS['id']
del RWS['goatPoints']
del RWS['recordHolders_0_active']
del RWS['recordHolders_0_country_name']

del RWS['recordHolders_1_active']
del RWS['recordHolders_1_country_name']

del RWS['recordHolders_2_active']
del RWS['recordHolders_2_country_name']
#print(RWS.head())

#Mascara 1
big3_RWS1 = RWS[(RWS['Record Holder 1'] == 'Roger Federer') | (RWS['Record Holder 1'] == 'Rafael Nadal') | (RWS['Record Holder 1'] == 'Novak Djokovic')]
#print(big3_RWS1)

#Mascara 2
big3_RWS2 = RWS[(RWS['Record Holder 2'] == 'Roger Federer') | (RWS['Record Holder 2'] == 'Rafael Nadal') | (RWS['Record Holder 2'] == 'Novak Djokovic')]
#print(big3_RWS2)

#Mascara 3
big3_RWS3 = RWS[(RWS['Record Holder 3'] == 'Roger Federer') | (RWS['Record Holder 3'] == 'Rafael Nadal') | (RWS['Record Holder 3'] == 'Novak Djokovic')]
#print(big3_RWS3)

#Se concatenan las mascaras
big3_RWS = pd.concat([big3_RWS1, big3_RWS2, big3_RWS3])
big3_RWS_final = big3_RWS.loc[:, 'name':'Detail 3']
big3_RWS_final
```

```python
big3_RWS_final.to_csv('big3_RWS_final.csv')
```

```python
big3_RWS_final.shape
```
